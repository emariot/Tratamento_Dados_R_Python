```r
# Maiores Empenhos
ME = head(data[order(-data$VALOREMPENHO),],10)
ME
```

```r
par(mar = c(7, 4, 4, 2) + 0.1)

barplot(ME$VALOREMPENHO, 
        names=ME$MUNICIPIO,
        cex.names=0.5,
        col=brewer.pal(n = 10, name = "RdBu"), 
        las=2, 
        main='Maiores Empenhos',
        #horiz=T,
       )
#legend("topright", legend=ME$MUNICIPIO, col = brewer.pal(n = 10, name = "RdBu"),lty=1:2, cex=0.8, ncol = 2,lwd=4)
box()
```

```r
# Maiores Pibs
MP = head(data[order(-data$PIB),], 10)
MP
```

```r
par(mar = c(7, 4, 4, 2) + 0.1)

barplot(MP$PIB, 
        names=MP$MUNICIPIO,
        cex.names=0.5,
        col=brewer.pal(n = 10, name = "RdBu"), 
        las=2, 
        main='Maiores PIBs'
        #horiz=T,
       )
box()
```

```r
# Rela√ß√£o entre PIB e Empenho
data$prop = data$VALOREMPENHO/data$PIB 
Prop = head(data[order(-data$prop),],10)
Prop
```

```r
par(mar = c(7, 4, 4, 2) + 0.1)

barplot(Prop$prop, 
        names=Prop$MUNICIPIO,
        cex.names=0.5,
        col=brewer.pal(n = 10, name = "RdBu"), 
        las=2, 
        main='Rela√ß√£o Empenho x PIB',
        #horiz=T,
       )
box()
```
