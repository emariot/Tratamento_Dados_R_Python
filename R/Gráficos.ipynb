* Histogramas: verificação da distribuição dos dados e frequência da variável


```r
head(trees)
```

```r
hist(trees$Height, main='Árvores', ylab='Frequência', xlab='Altura',  col='blue', density=20, breaks=20)
```

* Densidade

```r
densidade = density(trees$Height)
plot(densidade)
```

* Densidade sobre o histograma

```r
hist(trees$Height, main=NULL, xlab=NULL, ylab=NULL)
par(new=T) # Para sobrepor o próximo gráfico
plot(densidade)
```

* Dispersão: comparar variáveis contínuas

```r
plot(trees$Girth, trees$Volume)
```

```r
plot(trees$Girth, trees$Volume, main='Árvores', xlab='Volume', col='blue',)
# lementos gráficos
# type="l"
# pch=20
```

```r
# jitter: minimiza o efeito da sobreposição
plot(jitter(trees$Girth), trees$Volume, main='Árvores', xlab='Volume', col='blue',)
```

```r
split.screen(figs=c(2,2))
screen(1)
plot(trees$Girth, trees$Volume)
screen(2)
plot(trees$Girth, trees$Height)
screen(3)
plot(trees$Height, trees$Volume)
screen(4)
hist(trees$Volume)
close.screen(all=T)
```

* Boxplot: crve para compreender a distribuição dos dados

```r
boxplot(trees$Volume, main='Árvore', xlab='Volume')
```

```r
# Horizontal e sem ouline
boxplot(trees$Volume, main='Árvore', xlab='Volume', col='blue', horizontal=T, outline=F)
```

```r
boxplot(trees$Height, main='Árvore', xlab='Volume', notch=TRUE, horizontal=T)
```

```r
# Dados gerados peelo boxplot
boxplot.stats(trees$Height)
```

```r
boxplot.stats(trees$Height)$stats
```

```r
# Agregação 
head(InsectSprays)
```

```r
spray = aggregate(. ~ spray, data=InsectSprays, sum)
spray
```

* Gráfico de Barras

```r
barplot(spray$count, col=gray.colors(6),xlab='Spray',ylab='Total', names.arg=spray$spray)
box()
```

* Gráfico de Setor

```r
pie(spray$count, labels=spray$spray, main='Spray', col=c(1:6))
```

```r
# com legendas
pie(spray$count, labels=NA, main='Spray', col=c(1:6))
legend("bottomright", legend=spray$spray, cex=1, fill=c(1:6))
```
